```python
import pandas as pd
```

```python
# загрузка датасета
df = pd.read_csv('bookings.csv', sep=';')
df
```

```python
# изучение типов данных
df.dtypes
```

```python
# смотрю какие колонки в датафрейме
df.columns
```

```python
# привожу названия колонок к нижнему регистру и заменяю пробелы на нижнее подчеркивание
df = df.rename(columns = str.lower)
df = df.rename(columns=lambda col: col.replace(" ", "_"))
```

```python
df
```

```python
# Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

success_5 = df \
    .query("is_canceled == 0") \
    .groupby('country', as_index = False) \
    .aggregate({'is_canceled':'count'}) \
    .sort_values('is_canceled', ascending = False) \
    .head()
success_5
```

```python
# На сколько ночей в среднем бронируют отели разных типов?

nights = df \
    .groupby(['hotel'], as_index = False) \
    .agg({'stays_total_nights':'mean'})
nights.round(2)

#df.groupby('hotel').stays_total_nights.mean().round(2)
```

```python
# Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type). 
# Такое может произойти, например, по причине овербукинга. Сколько подобных наблюдений встретилось в датасете?

overbooking = df \
    .query("assigned_room_type != reserved_room_type").shape[0]
print("Фактически полученных номеров, отличающихся от забронированных", overbooking)

# len(df.loc[df.assigned_room_type != df.reserved_room_type])
```

```python
# На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017 году?

most_pop_month_2016 = df \
    .query("arrival_date_year == 2016") \
    .arrival_date_month.value_counts()
most_pop_month_2016.head(1)
```

```python
most_pop_month_2017 = df \
    .query("arrival_date_year == 2017") \
    .arrival_date_month.value_counts()
most_pop_month_2017.head(1)
```

```python
# Сгруппируйте данные по годам, 
# а затем проверьте, на какой месяц (arrival_date_month) бронирования отеля типа City Hotel 
#отменялись чаще всего в 2015? 2016? 2017? 

(
df
    .query("hotel == 'City Hotel' and is_canceled == 1 ")
    .groupby('arrival_date_year')
    .arrival_date_month
    .value_counts()
)
```

```python
# Посмотрите на числовые характеристики трёх колонок: adults, children и babies. 
# Какая из них имеет наибольшее среднее значение?

df[['adults', 'children', 'babies']].mean().head(1)
```

```python
# Создайте колонку total_kids, объединив столбцы children и babies
# Для отелей какого типа среднее значение переменной оказалось наибольшим?

df['total_kids'] = df.children + df.babies
(
    df
    .groupby('hotel')
    .agg({'total_kids': 'mean'})
    .round(2)
)


# df['total_kids'] = df['children'] + df['babies']
```

```python
# Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, сколько клиентов было потеряно в процессе. 
# Churn Rate.

df['has_kids'] = df['total_kids'] > 0
no_kids_churn = df.query("has_kids == False and is_canceled == 1").shape[0] / df.query("has_kids == False").shape[0]
no_kids_churn = round(no_kids_churn*100, 2)
chrn = '{}'
chrn = chrn.format(no_kids_churn)
print("Churn Rate для пользователей, которые не указали детей при бронировании", chrn)
```

```python
has_kids_churn = df.query("has_kids == True and is_canceled == 1").shape[0] / df.query("has_kids == True").shape[0]
has_kids_churn = round(has_kids_churn*100, 2)
chrn_2 = '{}'
chrn_2 = chrn_2.format(has_kids_churn)
print("Churn Rate для пользователей, которые указали детей при бронировании", chrn_2)
```
