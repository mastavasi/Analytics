```python
miss_data = df.isnull()
for column in miss_data.columns.values.tolist():
    print(column)
    print(miss_data[column].value_counts())
    print(" ")
# создал дф с нуллами по каждому столбцу, чтобы понимать количество пропущенных значений
# запустил цикл для подсчета пропущенных значений в каждом столбце
# уже можно сделать вывод о том, что нет пропущенных значений в айди и в датах (ошибок в данных нет)
# в полях DIGITAL_30_CNT TRAN_ACTIVE_30_CNT OPER_ACTIVE_30_CNT пропущено равное количество,
# значит в эти отчетные периоды пользователи не проявляли абсолютно никакой активности (и также нет ошибок)
```

```python
fill_na = df.fillna(0) 
# заполняю наны нулями, для удобства агрегации 
active_clients = fill_na \
    
active_clients = active_clients.drop(['DIGITAL_30_CNT', 'TRAN_ACTIVE_30_CNT', 'OPER_ACTIVE_30_CNT'], axis = 1)
active_clients['ACTIVE_TOTAL'] = active_clients['CLIENT_ID']
active_clients = active_clients\
    .groupby(['VALUE_DAY','CLIENT_ID'], as_index=False)\
    .agg({'ACTIVE_TOTAL':'count'})
 
active_clients
# количество всех клиентов за указанные периоды
```

```python
fill_na
```

```python
digital_active = fill_na \
    .query('DIGITAL_30_CNT != 0') \
    .groupby(['VALUE_DAY','CLIENT_ID'], as_index=False) \
    .agg({'DIGITAL_30_CNT':'sum'})
digital_active
```

```python
oper_active = fill_na \
    .query('OPER_ACTIVE_30_CNT != 0') \
    .groupby(['VALUE_DAY','CLIENT_ID'], as_index=False) \
    .agg({'OPER_ACTIVE_30_CNT':'sum'})
oper_active
```

```python
tran_active = fill_na \
    .query('TRAN_ACTIVE_30_CNT != 0') \
    .groupby(['VALUE_DAY','CLIENT_ID'], as_index=False) \
    .agg({'TRAN_ACTIVE_30_CNT':'sum'})
tran_active
```

```python
merge = active_clients \
    .merge(digital_active, on=['VALUE_DAY','CLIENT_ID'], how='left') \
    .merge(oper_active, on=['VALUE_DAY','CLIENT_ID'], how='left') \
    .merge(tran_active, on=['VALUE_DAY','CLIENT_ID'], how='left')

merge.head(60)
```

```python

segment_funnel = merge.drop('CLIENT_ID', axis = 1)

segment_funnel = segment_funnel.groupby('VALUE_DAY', as_index=False).count().sort_values('VALUE_DAY')
segment_funnel
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='VALUE_DAY', y='ACTIVE_TOTAL', data=segment_funnel, color='red', label='Total_Active')
sns.barplot(x='VALUE_DAY', y='DIGITAL_30_CNT', data=segment_funnel, color='blue', label='Digital')
sns.barplot(x='VALUE_DAY', y='OPER_ACTIVE_30_CNT', data=segment_funnel, color='green', label='Oper')
sns.barplot(x='VALUE_DAY', y='TRAN_ACTIVE_30_CNT', data=segment_funnel, color='orange', label='Tran')
plt.xlabel('VALUE_DAY')
plt.ylabel('Value')
plt.title('Bar Plot')
plt.legend(loc = 'lower right')
plt.xticks(rotation=45)
plt.show()
```
